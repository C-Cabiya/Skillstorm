### Getting rid of useless columns


```python
import pandas as pd 

movies = pd.read_csv("1000.csv")
movies =movies.drop(columns = ["Duration", "Description"])
movie = movies
```

### Oldest Movie

```python
years = movies["Title"].str.extract(r"([(]19\d\d[)]|[(]20\d\d[)])") # Extracting only the years in parentheses
# To avoid false matching 2049 from the Bladerunner film etc. 
years = years[0].str.extract(r"(19\d\d|20\d\d)") # Extracting the year from that already filteres Series
years = years[0].astype(int)
movies.loc[years == max(years)]
```

### Newest Movie

```python
movies.loc[years == min(years)].head(1) # Getting the first in the list of movies released in 1920, as there were more than one
# 1920 was the earliest year present in the data
```

### Top 10 Movies by IMDB Rating
Making the distinction from Metascore ratings, which are also present in the data; we will be using the 0-10 ratings

```python
#df.sort_values('Total',inplace=True,ascending=False)
movies.sort_values("Rate", ascending=False).head(10) # Top 10 when sorted by descending score
```

### Top Genres

```python
movies["Genre"] = movies["Genre"].str.split(",")
movies = movies.explode(column="Genre")
genre_ratings = movies.groupby("Genre").agg({"Rate":"max"})
genre_ratings.sort_values(by = "Rate", ascending=False) # Drama is the most highly rated genre
```

### Director with most movies

```python
people = movie["Cast"]
people = people.str.replace(r"Directors*:", "").str.strip()
people = people.str[10:].str.strip()
directors = people.str.replace(r"(\| (\w|\W)*)", "", regex=True).str.strip()
directors = directors.str.split(",")
df = directors.explode()
df.value_counts()
```

### Gross Revenue

```python
info = movie["Info"]
info = info.loc[info.str.contains("Gross")] # Only summing the values that actually have a value for gross revenue
info = info.str.replace (r"(\w|\W)* \| Gross: ", "", regex = True)
info = info.str.replace(r"$", "").str.replace("M","")
print(f"{info.astype(float).sum()} million dollars in gross revenue!")
```
